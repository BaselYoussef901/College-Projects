```python
import numpy as np
import scipy.io.wavfile as wavfile
import matplotlib.pyplot as plt
```

```python
h = np.zeros(99)
for i in range(99):
    h[i] = 0.31752 * np.sin(0.314159 * (i-49.00001)) / (i-49.00001)
    h[i] = h[i] * (0.54 - 0.46 * np.cos(0.0641114 * i))

plt.plot(h)
plt.title('Impulse Response')
plt.xlabel('n')
plt.ylabel('h[n]')
plt.show()
```

```python
x1 = np.zeros(500)
x1[0] = 1
y1 = np.convolve(x1, h)

plt.plot(y1)
plt.title('Output Signal 1')
plt.xlabel('n')
plt.ylabel('y1[n]')
plt.show()
```

```python
n = np.arange(500)
x2 = np.sin(2*np.pi*6*n/500) + 0.5*np.sin(2*np.pi*44*n/500)

plt.plot(x2)
plt.title('Input Signal 2')
plt.xlabel('n')
plt.ylabel('x2[n]')
plt.show()
```

```python
y2 = np.convolve(x2, h)

plt.plot(y2)
plt.title('Output Signal 2')
plt.xlabel('n')
plt.ylabel('y2[n]')
plt.show()
```
